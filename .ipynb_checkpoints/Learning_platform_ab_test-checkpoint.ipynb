В обоих группах отклоняем нулевую гипотезу что распределения нормальны. Для подтверждения статистической значимости увеличения ARPPU, сравним средние значения выборок,используя бутстрап 

```python
# Объявим функцию, которая позволит проверять гипотезы с помощью бутстрапа


def get_bootstrap(
    data_column_1,  # числовые значения первой выборки
    data_column_2,  # числовые значения второй выборки
    boot_it=1000,  # количество бутстрэп-подвыборок
    statistic=np.mean,  # интересующая нас статистика
    bootstrap_conf_level=0.95  # уровень значимости
):
    boot_len = max([len(data_column_1), len(data_column_2)])
    boot_data = []
    for i in tqdm(range(boot_it)):  # извлекаем подвыборки
        samples_1 = data_column_1.sample(
            boot_len,
            replace=True  # параметр возвращения
        ).values

        samples_2 = data_column_2.sample(
            boot_len,
            replace=True
        ).values

        # mean() - применяем статистику
        boot_data.append(statistic(samples_1-samples_2))

    pd_boot_data = pd.DataFrame(boot_data)

    left_quant = (1 - bootstrap_conf_level) / 2
    right_quant = 1 - (1 - bootstrap_conf_level) / 2
    ci = pd_boot_data.quantile([left_quant, right_quant])

    p_1 = ss.norm.cdf(
        x=0,
        loc=np.mean(boot_data),
        scale=np.std(boot_data)
    )
    p_2 = ss.norm.cdf(
        x=0,
        loc=-np.mean(boot_data),
        scale=np.std(boot_data)
    )
    p_value = min(p_1, p_2) * 2

    # Визуализация
    plt.hist(pd_boot_data[0], bins=50)

    plt.style.use('ggplot')
    plt.vlines(ci, ymin=0, ymax=50, linestyle='--')
    plt.xlabel('boot_data')
    plt.ylabel('frequency')
    plt.title("Histogram of boot_data")
    plt.show()

    return {"boot_data": boot_data,
            "ci": ci,
            "p_value": p_value}
```

```python
get_bootstrap(
    data_column_1=test_data_df.query('grp == "B" & rev>0').rev,
    data_column_2=test_data_df.query('grp == "A" & rev>0').rev,
    statistic=np.mean
)
```

Значение p-value меньше 0.05, что позволяет нам отклонить нулевую гипотезу о равенстве средних значений

```python
# Создадим датафрейм для метрик

df_for_plot = pd.DataFrame({'CR_a': [],
                            'CR_b': [],
                            'ARPPU_a': [],
                            'ARPPU_b': [],
                            'ARPU_a': [],
                            'ARPU_b': []
                            })
```

```python
# Объявим функцию


def get_update(
    path  # путь к обновленной таблице
):
    add_info_df = pd.read_csv(path)  # считываем данные для обновления
    # Обьединяем старые и новые данные
    control_test_df_update = pd.concat([control_test_df, add_info_df])
    # Создаём объединённый датафрейм
    ab_info_df = full_data_df.merge(
        control_test_df_update, left_on='student_id', right_on='id')
    ab_info_df.drop('id', axis=1, inplace=True)  # удалим лишнюю колонку
    # добавим колонку с бинарной переменной для конверсии
    ab_info_df['converted'] = ab_info_df['rev'].apply(
        lambda x: 1 if x > 0 else 0)
    # Расчитаем метрики
    CR_a = ab_info_df.query('grp == "A"').converted.mean()
    CR_b = ab_info_df.query('grp == "B"').converted.mean()
    ARPPU_a = ab_info_df.query('grp == "A" & rev>0').rev.mean()
    ARPPU_b = ab_info_df.query('grp == "B" & rev>0').rev.mean()
    ARPU_a = CR_a * ARPPU_a
    ARPU_b = CR_b * ARPPU_b
    # Создадим строку для датафрейма метрик
    new_row = {
        'CR_a': CR_a,
        'CR_b': CR_b,
        'ARPPU_a': ARPPU_a,
        'ARPPU_b': ARPPU_b,
        'ARPU_a': ARPU_a,
        'ARPU_b': ARPU_b
    }
    # Обновим датафрейм с метриками значениями которые посчитала функция
    global df_for_plot
    df_for_plot = df_for_plot.append(new_row, ignore_index=True)
    return df_for_plot
```

```python
# Записываем в переменную путь к дополнительным данным

path = 'https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/usRqf-Qs2XMJeA'
```

```python
# Используем функцию и обновим данные.
# Нижняя строка датафрейма-новые значения метрик.

get_update(path)
```

```python
# Пишем функцию в которую будем передавать
# сформированный прошлой функцией датафрейм.


def get_plot(df):
    df = df.reset_index()
    plt.subplot(2, 3, 1)
    lp = sns.lineplot(y=df.CR_a, x=df.index)
    lp.axes.set_title("Конверсия контроль", fontsize=16)
    lp.set_xlabel("Этап", fontsize=14)
    lp.set_ylabel("Значения", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(w_pad=2)

    plt.subplot(2, 3, 2)
    lp = sns.lineplot(y=df.ARPPU_a, x=df.index)
    lp.axes.set_title("ARPPU контроль", fontsize=16)
    lp.set_xlabel("Этап", fontsize=14)
    lp.set_ylabel("Значения", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(w_pad=2)

    plt.subplot(2, 3, 3)
    lp = sns.lineplot(y=df.ARPU_a, x=df.index)
    lp.axes.set_title("ARPU контроль", fontsize=16)
    lp.set_xlabel("Этап", fontsize=14)
    lp.set_ylabel("Значения", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(w_pad=2)

    plt.subplot(2, 3, 4)
    lp = sns.lineplot(y=df.CR_b, x=df.index)
    lp.axes.set_title("Конверсия тест", fontsize=16)
    lp.set_xlabel("Этап", fontsize=14)
    lp.set_ylabel("Значения", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(w_pad=2)

    plt.subplot(2, 3, 5)
    lp = sns.lineplot(y=df.ARPPU_b, x=df.index)
    lp.axes.set_title("ARPPU тест", fontsize=16)
    lp.set_xlabel("Этап", fontsize=14)
    lp.set_ylabel("Значения", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(w_pad=2)

    plt.subplot(2, 3, 6)
    lp = sns.lineplot(y=df.ARPU_b, x=df.index)
    lp.axes.set_title("ARPU тест", fontsize=16)
    lp.set_xlabel("Этап", fontsize=14)
    lp.set_ylabel("Значения", fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(w_pad=2)
    return plt.show()
```

```python
# Запускаем функцию
get_plot(df_for_plot)
```

Основной идеей было получить некое подобие приборной панели нашего эксперимента. На имеющихся у нас двух этапах визуализация далека от идеальной, скорее сейчас она показывает вектор в котором движутся метрики но если чаще обновлять данные, будет более наглядно.
