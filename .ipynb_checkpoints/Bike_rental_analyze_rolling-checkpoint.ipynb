## Поиск аномалий в  данных о аренде велосипедов в Лондоне
В нашем распоряжении датасет с данными  за два два года: с 4 января 2015 по 3 января 2017.
Изучим динамику числа аренд, наличие связи с погодными условиями и выходными, поищем аномалии в данных и найдём им объяснение.

```python
#Импортируем библиотеки
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Определяем настройки отображения графиков
sns.set(
    font_scale=2,
    style="whitegrid",
    rc={'figure.figsize':(20,8)}
        )
```

```python
#Считаем датафрейм
bike = pd.read_csv('C:/Users/79256/Github/Data_Analyse/Data/london_bike_rental.csv',parse_dates=['timestamp'])
bike.head()
```

```python
#Проверим количество строк и столбцов
bike.shape
```

```python
 #Проверим количество пропущенных значений
bike.isna().sum()
```

```python
#И типы данных
bike.dtypes
```

```python
#Построим график с датами и количеством аренд. График не наглядный, нужно
#агрегировать количество аренд по дням 
sns.lineplot(x=bike.timestamp, y=bike.cnt)
```

```python
#При помощи resample агрегируем данные по дням
bike_day = bike.sort_values('timestamp') \
    .set_index('timestamp') \
    .resample(rule='D') \
    .sum()
```

```python
bike_day.head(3)
```

```python
#Визуализируем количество аренд по дням
sns.lineplot(x=bike_day.index, y=bike_day.cnt)
```

```python
#Добавим столбец с трехдневным скользящим средним
bike_day['MA_3'] = bike_day.cnt.rolling(3) \
                            .mean()
```

```python
#И построим график со скользящей средней
sns.lineplot(x=bike_day.index, y=bike_day.cnt)
sns.lineplot(x=bike_day.index, y=bike_day.MA_3)
```

```python
#Сбросим колонку timestamp из индекса обратно в колонку
bike_day.reset_index(inplace=True)
```

```python
#Вычислим разницу между значениями количества аренд и значениями скользящего
#среднего(результат будет Пандас серия)
c = bike_day.cnt - bike_day.MA_3
```

```python
#Вычислим стандартное отклонение полученной серии
std = np.std(c)
std
```

```python
#Добавим колонки с верхней и нижней границей 99% доверительного интервала
bike_day['upper_bound'] = bike_day['MA_3'] + 2.576 * std
bike_day['lower_bound'] = bike_day['MA_3'] - 2.576 * std
```

```python
bike_day.head(3)
```

```python
#Отберём дни когда значения количества аренд превысили верхнюю границу интервала
bike_day.query('cnt > upper_bound ') \
        .sort_values('cnt', ascending=False)
```

### Изучаем новости и приходим к выводу, что причина аномального количества аренд 09.07.2015 в забастовке работников метро

```python
#Отберём дни когда значения количества аренд ниже нижней границы интервала
bike_day.query('cnt < upper_bound ') \
        .sort_values('cnt')
```

### 02.09.2016 во всех колонках нули, таким образом причина самого низкого количества аренд в датафрейме - ошибка при вводе данных
