## Поиск возможных причин оттока пользователей - водителей такси
Есть данные о такси-компании, которая хочет изучить отток водителей и посмотреть, какие есть различия между водителями, которые покидают сервис, и которые остаются. Нужно сформулировать и протестировать гипотезы, выделить группы водителей, которые наиболее подвержены "оттоку". На основе результатов сделать выводы о том, что можно улучшить в сервисе, чтобы в дальнейшем внести изменения.
Еще раз список переменных:

- `city` – город
- `phone` – основное устройство, которое использует водитель 
- `signup_date` – дата регистрации аккаунта (`YYYYMMDD`)
- `last_trip_date` – дата последней поездки (`YYYYMMDD`)
- `avg_dist` – среднее расстояние (в милях) за поездку в первые 30 дней после регистрации
- `avg_rating_by_driver` – средняя оценка поездок водителем 
- `avg_rating_of_driver` – средняя оценка поездок водителя
- `surge_pct` – процент поездок, совершенных с множителем > 1 
- `avg_surge` – средний множитель всплеска за все поездки этого водителя
- `trips_in_first_30_days` – количество поездок, которые совершил водитель в первые 30 дней после регистрации
- `luxury_car_user` – TRUE, если пользователь в первые 30 дней использовал премиум-автомобиль
- `weekday_pct` – процент поездок пользователя, совершенных в будние дни

## графики, гипотезы и тесты

Создадим колонку churn с данными о том, ушел ли пользователь
 как baseline принимаем гипотезу, что пользователь ушел, если не был активен последние __30__ дней

```python
#Посмотрим последнюю дату, за которую есть информация о поездках. Примем её за последнюю известную дату
df.last_trip_date.max()
```

```python
#Добавим колонку с количеством дней между последней известной датой и последней поездкой
df['days_since_last_trip'] = df.last_trip_date.max() - df.last_trip_date
```

```python
#Колонка получила тип данных timedelta, для простоты расчетов преобразуем в int
df['days_since_last_trip']
```

```python
df['days_since_last_trip'] = df['days_since_last_trip'].dt.days
```

```python
df['days_since_last_trip']
```

```python
#Добавим колонку churn с бинарной переменной, ушел или нет водитель
df['churn'] = df.days_since_last_trip.apply(lambda x: 'churn' if x > 30 else 'not_churn')
df[['days_since_last_trip', 'churn']]
```

```python
#Видим, что по нашей метрике, приложение покинуло более 60% водителей
df.churn.value_counts(normalize=True).mul(100)
```

### Выдвинем гипотезы о причинах ухода

### churn & phone

Предположим, что проблема может быть среди юзеров на конкретной платформе:

```python
#Построим таблицу сопряжения
pd.crosstab(df.churn, df.phone)
```

```python
fig = px.histogram(df[['churn', 'phone']].dropna(), x='churn', 
                   color='phone', color_discrete_sequence=px.colors.sequential.Cividis)
fig.show()
```

По графику видно, что большинство покинувших приложение - пользователи Android.
Проверим нашу гипотезу с помощью статистического теста.

Так как обе переменные (phone и churn) категориальные, используем хи-квадрат
__нулевая гипотеза - взаимосвязи между переменными нет__

```python
from scipy.stats import chi2_contingency, chi2 
```

```python
stat, p, dof, expected = chi2_contingency(pd.crosstab(df.churn, df.phone))
```

```python
stat, p
```

Интерпретируем результат:

```python
prob = 0.95
critical = chi2.ppf(prob, dof)
if abs(stat) >= critical:
    print('Отклоняем H0')
else:
    print('Не отклоняем H0')
```

```python
prob = 0.95
alpha = 1.0 - prob
if p <= alpha:
    print('Отклоняем H0')
else:
    print('Не отклоняем H0')
```

Проверим распределения на нормальность при помощи формулы Шапиро-Уилка. __Нулевая гипотеза, что распределение нормально__

```python
ss.shapiro(
            df[df.city == "King's Landing"] \
                 .trips_in_first_30_days \
                 .sample(1000)
            )
```

```python
ss.shapiro(
            df[df.city == 'Astapor'] \
                 .trips_in_first_30_days \
                 .sample(1000)
            )
```

```python
ss.shapiro(
            df[df.city == 'Winterfell'] \
                 .trips_in_first_30_days \
                 .sample(1000)
            )
```

Так как количество групп больше двух и распределение отличное от нормального, используем непараметрический аналог ANOVA- критерий Краскела — Уоллиса __нулевая гипотеза между выборками существуют лишь случайные различия по уровню исследуемого признака__

```python
from scipy.stats import chi2_contingency, chi2, mannwhitneyu, shapiro, kruskal
import statsmodels.api as sa 
import scikit_posthocs as sp  # sp.posthoc_dunn() для непараметрического аналога anova 
```

```python
stat_k, p_k = kruskal(
                df[df.city == "King's Landing"].trips_in_first_30_days,
                df[df.city == 'Astapor'].trips_in_first_30_days,
                df[df.city == 'Winterfell'].trips_in_first_30_days
                )
```

```python
print(f'Statistik = {stat_k}, p-value = {p_k}')
```

```python
8.724567791938856e-49 < 0.05
```

Есть основания отклонить нулевую гипотезу, то есть между выборками есть статистически значимые различия

### trips_in_first_30_days & churn

Исследуем связь оттока водителей и активности в первые 30 дней с момента регистрации :

```python
df.head()
```

```python
#Посмотрим на средние и медианные значения активности в первые 30 дней для покинувших приложение и нет
#можно предположить, что ушедшие были менее активны
df.groupby('churn', as_index=False) \
  .agg({'trips_in_first_30_days':['mean','median']})
```

```python
#Построим графики распределений в разрезе оттока.
#Видим, что распределения ненормальны
sns.distplot(df.query('churn =="churn"').trips_in_first_30_days,kde=False)
sns.distplot(df.query('churn =="not_churn"').trips_in_first_30_days,kde=False)
```

Проверим тестом Шапиро-Уилка

```python
ss.shapiro(
            df[df.churn == 'churn'] \
                 .trips_in_first_30_days \
                 .sample(1000)
            )
```

```python
ss.shapiro(
            df[df.churn == 'not_churn'] \
                 .trips_in_first_30_days \
                 .sample(1000)
            )
```

Так как распределение ненормально и нужно сравнить два значения, выбираем непараметрический критерий Манна-Уитни
__Нулевая гипотеза - уровень признака во второй выборке не ниже уровня в первой__ 

```python
mannwhitneyu(
            df[df.churn == 'churn'].trips_in_first_30_days,
            df[df.churn == 'not_churn'].trips_in_first_30_days
            )
```

Получаем значение p-value меньше 0.05, что даёт основания отклонить нулевую гипотезу и сделать вывод о том, что ушедшие пользователи были менее активны
