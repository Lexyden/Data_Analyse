```python
#Для вычисления биномиального коэффиициента импортируем формулу факториала
from math import factorial
```

__Формула Байеса__ применяется при вопросах типа "В команде 20 стрелков, 1 группа- 15 человек попадает в мишень с вероятностью 0.6, вторая группа- 5 человек с вероятностью 0.8. Если случайный стрелок попадает в цель, какова вероятность, что он из первой группы.
Формула:  
P(Y|X) = P(Y) * P(X|Y) / P(X)  
Y-гипотеза(первая группа)  
X-условие(попал в цель)  
P- вероятность  
P(Y|X)- вероятность того, что стрелок из первой группы, если он попал в цель  
P(Y)-вероятность, что он из первой группы  
P(X|Y)- вероятность, что стрелок из первой группы попал в цель(дана в условии)  
P(X) - вероятность попадания = вероятность того, что из первой группы * вероятность попадания + вероятность того, что из второй группы * вероятность попадания.


### Задание 1

Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй - 2%, в третьей – 6%. Наудачу выбирается один компьютер. Определить вероятность того, что выбранный компьютер – бракованный.  Округлите полученный ответ до десятитысячных. В качестве разделителя используйте точку.

```python
'''
Перемножаем вероятность что принадлежит к группе 
на вероятность брака и складываем для каждой группы
'''
round(180/500 * 0.03 + 170/500 * 0.02 + (500-180-170)/500 * 0.06,4)
```

### Задание 3

Из 40 снайперов 18 попадает в цель с вероятностью 0,9, 8 - с вероятностью 0,4 и 14 – с вероятностью 0,7. Наудачу выбранный снайпер произвел выстрел, поразив цель. Мы хотим вычислить, снайпер из какой группы наиболее вероятно совершил этот выстрел. 

В ответ запишите наибольшую вероятность, округленную до тысячных. В качестве разделителя используйте точку.

```python
'''Расчитаем по формуле Байеса вероятности для каждой группы и сравним'''
p_1 = (18/40 * 0.9) / (18/40 * 0.9 + 8/40 * 0.4 + 14/40 * 0.7)
p_2 = (8/40 * 0.4) / (18/40 * 0.9 + 8/40 * 0.4 + 14/40 * 0.7)
p_3 = (14/40 * 0.7) / (18/40 * 0.9 + 8/40 * 0.4 + 14/40 * 0.7)
list=[p_1,p_2,p_3]
for i in list:
    print(round(i,3))
#Самая большая вероятность, что снайпер из первой группы
```

### Задание 4

Есть 4 кубика. На трех из них окрашена белым половина граней, а на четвертом кубике всего одна грань из шести белая. Наудачу выбранный кубик подбрасывается семь раз. Найти вероятность того, что был выбран четвертый кубик, если при семи подбрасываниях белая грань выпала ровно один раз.  Ответ округлите до тысячных. Используйте точку в качестве десятичного разделителя.

```python
'''Здесь нужно применить формулу Бернулли, чтоб расчитать вероятность выпадения
одной белой грани при семи броскаx. 
Пусть H1 это событие, когда выбирают 1, 2 или третий кубик,
а H2 - когда выбирают 4-ый кубик.'''
#Создадим функцию для вычисления биномиального коэффициента
def get_binomial(n,k):
    return factorial(n) / (factorial(k) * factorial(n-k))
```

```python
def get_p_bernulli(n,k,p,q):
    p_bernulli = get_binomial(n,k) * p**k * q**(n-k)
    return p_bernulli
```

```python
p_h1 =round(get_p_bernulli(n=7,k=1,p=1/2,q=1/2),4)
p_h1
```

```python
p_h2 =round(get_p_bernulli(n=7,k=1,p=1/6,q=5/6),4)
p_h2
```

```python
p_cube = 1/4 * p_h2 / (1/4 * p_h2 + 3/4* p_h1) 
```

```python
round(p_cube,3)
```
