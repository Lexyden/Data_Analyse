Тренировка по построению линейной регрессии на основе тестового OZONE camp (проверка решения в EXCEL)

```python
#Импортируем библиотеки
from sklearn.datasets import load_boston
import pandas as pd
import seaborn as sns
import numpy as np
%matplotlib inline
from math import sqrt
```

```python
#Определяем настройки отображения графиков
sns.set(
    font_scale=2,
    style="whitegrid",
    rc={'figure.figsize':(20,8)}
        )
```

```python
#Сформируем датасет по данным из тестового
lr_df = pd.DataFrame({'a':[100,230,280,320,330,400,470,510],'b':[12,19,25,40,35,50,52,60]})
```

```python
lr_df
```

```python
#Визуализируем регрессию
sns.regplot(data=lr_df,x='a',y='b')
```

```python
#Построим линейную регрессию через statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
results = smf.ols('b ~ a',lr_df).fit()
print(results.summary())
```

```python
#Подготовим список независимых переменных из задания чтоб по ним определить зависимые
test_list_a=np.array([150,200,450]).reshape(-1, 1)
```

```python
#Импортируем модель линейной регрессии из sklearn 
from sklearn.linear_model import LinearRegression
```

```python
#Обучим модель на даннных из задания.В fit нужно подавать двумерные массивы
model_for_excel = LinearRegression().fit(lr_df.a.values.reshape(-1,1),lr_df.b.values.reshape(-1,1))
```

```python
result_a_predict = model_for_excel.predict(test_list_a)
result_a_predict.round()
```
