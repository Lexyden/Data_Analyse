## Рассмотрим визуализацию корреляции и простой линейной регрессии

#### Корреляция
Коэффициент корреляции в своём статистическом смысле обозначает силу и характер взаимосвязи между двумя количественными переменными. Взаимосвязь может быть положительной (когда одна переменная растёт, другая тоже растёт), либо отрицательной (когда одна переменная растёт, другая уменьшается), либо отсутствовать.

```python
from sklearn.datasets import load_boston
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats
%matplotlib inline
from math import sqrt
```

```python
sns.set(
    font_scale=2,
    style="whitegrid",
    rc={'figure.figsize':(20,8)}
        )
#Определяем настройки отображения графиков
```

```python
#Загрузим учебный датасет о размере чаевых
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
#Визуализируем наблюдения при помощи scutterplot
sns.scatterplot(data=tips_df, x='total_bill', y='tip', hue='size', size='size')
```

```python
'''Один из методов подсчета коэффициента корреляции - метод corrcoef в библиотеке
numpy. Он использует параметрический критерий Пирсона и не считает p-value'''
np.corrcoef(tips_df.total_bill, tips_df.tip)
```

```python
'''Второй метод из библиотеки stats. Есть параметрический критерий Пирсона
но также возвращает p-value'''
scipy.stats.pearsonr(tips_df.total_bill, tips_df.tip)
```

#### Ограничения использования параметрического критерия:
- связь линейна и монотонна
- отсутствуют сильные выбросы
- переменные нормально распределены

```python
'''Аналогично используем непараметрический критерий Спирмена'''
scipy.stats.pearsonr(tips_df.total_bill, tips_df.tip)
```
